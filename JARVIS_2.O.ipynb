```python
import pyttsx3
import datetime
import webbrowser
import os
import pyautogui
import speech_recognition as sr
from langchain_ollama import OllamaLLM
from langchain_core.prompts import ChatPromptTemplate

engine=pyttsx3.init('sapi5')
voices=engine.getProperty('voices')
#print(voices)
engine.setProperty('voice', voices[0].id)


template="""
     Answer the question below.

     Here is the conversation history: {context}

     Question={question}

     Answer: 
     """
model=OllamaLLM(model="llama3")
prompt=ChatPromptTemplate.from_template(template)
chain=prompt|model


def speak(audio):
    engine.say(audio)
    engine.runAndWait()


def wishMe():
    print("...JARVIS ONLINE...")
    speak("JARVIS ONLINE!")
    hour=int(datetime.datetime.now().hour)
    if hour>=0 and hour<12:
        speak("Good Mornig Sir!!")
    elif hour>=12 and hour<18:
        speak("Good afternoon Sir!!")
    else:
        speak("Good Evening Sir!!")
    speak("How can I help You?")


def takecommand():
    r=sr.Recognizer()
    with sr.Microphone() as source:
         print("JARVIS:Listening...")
         r.pause_threshold=3
         audio=r.listen(source)
    try:
         print("JARVIS:Recognising...")
         query=r.recognize_google(audio, language='en-in')
         print("User said: ", query)
    except Exception as e:
         print(e)

         print("JARVIS:Sir!!!  Can you repeat that again.... please!")
         speak("Sir!!!  Can you repeat that again.... please!")
         return None
    return query


def chatbot(user_input,context):
     result=chain.invoke({"context": context, "question": user_input})
     print("JARVIS: \n", result) 
     speak(result)
     context = (context or "") + f"\nUser::{user_input}\nAI:{result}"  
     return context


def performcommand(query):
    if 'open youtube' in query:
         speak("Opening youtube for you sir!")
         webbrowser.open("youtube.com")

    elif 'close youtube' in query:
          speak('closing youtube')
          pyautogui.hotkey("ctrl","w")


    elif 'open google' in query:
         speak("Opening google for you sir!")
         webbrowser.open("google.com")  
    
    elif 'close google' in query:
          speak('closing google')
          pyautogui.hotkey("ctrl","w")

    elif 'open hi anime' in query or 'high anime' in query or 'anime' in query:
         speak("Opening hi anime for you sir!")
         webbrowser.open("https://hianime.to/home")
    
    elif 'close anime' in query:
          speak('closing anime')
          pyautogui.hotkey("ctrl","w")

    elif 'play songs' in query or 'play song' in query or 'play music' in query:
         speak("Playing songs from 'E' directory songs folder...  sir !")
         songs_dir="E:\\Songs"
         songs=os.listdir(songs_dir)
         print(songs)
         os.startfile(os.path.join(songs_dir,songs[0]))
        
    elif 'time' in query:
         strtime=datetime.datetime.now().strftime("%H:%M:%S")
         print(strtime)
         speak(f"Sir, The time is {strtime}")
        
    elif 'date' in query:
         strdate=datetime.datetime.now().strftime("%Y-%m-%d")
         print(strdate)
         speak(f"Sir,Todays date is {strdate}")
        
    elif 'open vs code' in query:
         speak("Opening Visual Studio for you sir!")
         code_path="D:\\apps\\Microsoft VS Code\\Code.exe"
         os.startfile(code_path)
        
    elif 'open whatsapp' in query:
         speak("Opening your whats app account sir!")
         whatsapp_path="explorer shell:AppsFolder\\5319275A.WhatsAppDesktop_cv1g1gvanyjgm!App"
         os.system(whatsapp_path)
    
    elif 'close whatsapp' in query or 'exit whatsapp' in query:
         speak("Closing WhatsApp for you, sir!")
         os.system("taskkill /F /IM WhatsApp.exe")
   
    elif 'open instagram' in query:                              
         speak("Opening your instagram sir!")
         Instagram_path="explorer shell:AppsFolder\\Facebook.InstagramBeta_8xx8rvfyw5nnt!App"
         os.system(Instagram_path)
    
    elif 'close instagram' in query or 'exit instagram' in query:
         speak("Closing instagram for you, sir!")
         os.system("taskkill /F /IM Instagram.exe")

    elif 'open spotify' in query:
          speak("Opening spotify for you sir! Enjoy your music!")
          Spotify_path='start spotify'
          os.system(Spotify_path)
      
    elif 'close spotify' in query or 'exit spotify' in query:
         speak("Closing spotify for you, sir!")
         os.system("taskkill /F /IM Spotify.exe")
     
     
     
if __name__=="__main__":
    speak("greetings!")
    wishMe()
    context=""
        
    text=["open","close","open youtube","open google","open whatsapp","open spotify","open hi anime","open instagram","open vs code","todays date","today's date","what is the time now?","time now","play"]
        
    while (True):
     query=takecommand()
     if query:
         query=query.lower()
         if any(word in query for word in text):
                performcommand(query)
         elif 'exit' in query or 'shutdown' in query or 'terminate' in query:
                  speak("Thank You Sir! Have A Nice day! Shutting Down initialized....")
                  speak("Going offline!")
                  print("...JARVIS IS OFFLINE NOW...")
                  break  
         else:
                  context=chatbot(query,context)
```
